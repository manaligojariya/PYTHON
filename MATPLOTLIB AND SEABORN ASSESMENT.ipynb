```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.get_dataset_names()
```

```python
abc = sns.load_dataset('iris')
abc
```

```python
# A. Creating and customizing line charts using Matplotlib 

plt.plot(abc)
plt.title('Sepal Length')
plt.xlabel('Sample Index')
plt.ylabel('Sepal Length')
plt.show()
```

```python
# B. Visualizing relationships between two or more variables using scatter plots
plt.figure(figsize=(10, 6))
sns.scatterplot(data=abc)
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.show()
```

```python
# C. Studying distributions of variables using histograms & bar charts to 
plt.hist(abc,bins=20)
plt.show()
```

```python
# D. Visualizing two-dimensional data using heatmaps
plt.figure(figsize = (15,8))
# del abc['species']
sns.heatmap(abc.corr(),annot=True,linewidth=.5)
plt.show()
```

```python
# E. Displaying images using Matplotlib's plt.imshow 

image = np.random.rand(10, 10)
plt.imshow(image, cmap='viridis')
plt.title('Sample Image')
plt.axis('off')
plt.show()
```

```python
# F. Plotting multiple Matplotlib and Seaborn charts in a grid 
sns.pairplot(abc)
plt.show()
```
