# Scores

```python
#Step 1. Import the necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Step 2. Create the DataFrame that should look like the one below.

# Creating the DataFrame with the given data
data = {
    'first_name': ['Jason', 'Molly', 'Tina', 'Jake', 'Amy'],
    'last_name': ['Miller', 'Jacobson', 'Milner', 'Ali', 'Cooze'],
    'age': [42, 52, 36, 24, 73],
    'female': [0, 1, 1, 0, 1],
    'preTestScore': [25, 24, 31, 2, 3],
    'postTestScore': [94, 57, 62, 70, 70]
}

df = pd.DataFrame(data)
print(df)
```

```python
# Step 3. Create a Scatterplot of preTestScore and postTestScore, with the size of each point determined by age
plt.figure(figsize=(10, 6))
plt.scatter(df['preTestScore'], df['postTestScore'], s=df['age']*10, alpha=0.5)

# Add labels and title
plt.xlabel('Pre-Test Score')
plt.ylabel('Post-Test Score')
plt.title('Scatter Plot of Pre-Test and Post-Test Scores')

# Show the plot
plt.show()
```

```python
#Step 4. Create a Scatterplot of preTestScore and postTestScore
# Create the scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(df['preTestScore'], df['postTestScore'], s=df['age'], alpha=0.5)

# Add labels and title
plt.xlabel('Pre-Test Score')
plt.ylabel('Post-Test Score')
plt.title('Scatter Plot of Pre-Test and Post-Test Scores')

# Show the plot
plt.show()
```

```python
#BONUS: Create your own question and answer it.
#How can we calculate the average postTestScore for males and females?
average_postTestScore_male = df[df['female'] == 0]['postTestScore'].mean()
average_postTestScore_female = df[df['female'] == 1]['postTestScore'].mean()

average_postTestScore_male, average_postTestScore_female
```

# APPLE STOCK

```python
# Step 1: Import the necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Step 2: Import the dataset from the given address
apple = pd.read_csv( "C:\\Users\\lenovo\\Documents\\Downloads\\apple_stock_data.csv")
apple
```

```python
# Step 4: Check out the type of the columns
column_types = apple.dtypes
column_types
```

```python
# Step 5: Transform the Date column as a datetime type
apple['Date'] = pd.to_datetime(apple['Date'])
apple
```

```python
# Step 7: Check for any duplicate dates
duplicates = apple.index.duplicated().sum()
duplicates
```

```python
# Step 8: Make the first entry the oldest date
apple.sort_index(inplace=True)
apple
```

```python
# Step 9: Get the last business day of each month
last_business_days = apple.resample('BM').last()
last_business_days
```

```python
# Step 10: Calculate the difference in days between the first day and the oldest
date_difference = (apple.index[-1] - apple.index[0]).days
date_difference
```

```python
# Step 11: Count the number of months in the data
num_months = apple.resample('M').size().shape[0]
num_months
```

```python
# Step 12: Plot the 'Adj Close' value
plt.figure(figsize=(13.5, 9))
plt.plot(apple['Adj Close'])
plt.title('Apple Stock Adjusted Close Price')
plt.xlabel('Date')
plt.ylabel('Adjusted Close Price')
plt.show()
```
